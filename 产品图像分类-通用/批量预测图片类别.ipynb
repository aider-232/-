```python
from 预测单张脚垫类图片的类别 import predict_from_url
from 预测单张脚垫类图片的类别 import load_model_and_classes
import pandas as pd
```

```python
df = pd.read_excel("asin和图片链接_测试.xlsx")
```

```python
a = 0
temp = []
model, class_names = load_model_and_classes('产品图像分类-通用.h5','产品图像分类_通用_类别.json')
for i in df['商品主图']:
    a = a + 1
    url_result, url_confidence = predict_from_url(i, model, class_names)
    temp.append(url_result)
    print(f"预测类别：{url_result}")
    print(f"置信度：{url_confidence:.2f}%")
    print(a)
```

```python
df['预测类型'] = temp
df.to_excel("asin和图片链接及预测类型_测试.xlsx")
```
