```python
from 预测单张脚垫类图片的类别 import predict_from_url
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Administrator\Desktop\脚垫类产品分类器\asin和图片链接.xlsx")
```

```python
a = 0
temp = []
for i in df['商品主图']:
    a = a + 1
    url_result, url_confidence = predict_from_url(i)
    temp.append(url_result)
    print(f"预测类别：{url_result}")
    print(f"置信度：{url_confidence:.2f}%")
    print(a)
```

```python
df['预测类型'] = temp
df.to_excel("asin和图片链接及预测类型.xlsx")
```
